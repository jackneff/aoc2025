# IMPORT

```python
def import_lines(file):
    with open(file, 'r') as f:
        lines = [line.rstrip() for line in f.readlines()]
        print(f"{file} has {len(lines)} lines")
        return lines
EXAMPLE_LINES = import_lines('ex.txt')
INPUT_LINES = import_lines('input.txt')
```

# PARSE

```python
SAMPLE_LIMIT = 5
def parse(data):
    return data

EXAMPLE_DATA = parse(EXAMPLE_LINES)
INPUT_DATA = parse(INPUT_LINES)
print(f"Example Data Top {SAMPLE_LIMIT}:")
print(EXAMPLE_DATA[:SAMPLE_LIMIT])
print(f"Input Data Top {SAMPLE_LIMIT}: ")
print(INPUT_DATA[:SAMPLE_LIMIT])
```
