# IMPORT

```python
def import_lines(file):
    with open(file, 'r') as f:
        lines = [line.rstrip() for line in f.readlines()]
        print(f"{file} has {len(lines)} lines")
        return lines
EXAMPLE_LINES = import_lines('ex.txt')
INPUT_LINES = import_lines('input.txt')
```

# PARSE

```python
SAMPLE_LIMIT = 5
def parse(data):
    return data

EXAMPLE_DATA = parse(EXAMPLE_LINES)
INPUT_DATA = parse(INPUT_LINES)
print(f"Example Data Top {SAMPLE_LIMIT}:")
print(EXAMPLE_DATA[:SAMPLE_LIMIT])
print(f"Input Data Top {SAMPLE_LIMIT}: ")
print(INPUT_DATA[:SAMPLE_LIMIT])
```

# PART 1

```python
answer = 0
for bank in INPUT_DATA:
    sv = str(bank)
    first = "1"
    second = "1"
    for i,v in enumerate(bank):
        #print(i,len(bank))
        if i != len(bank)-1 and v > first:
            second = "1"
            first = v 
        elif v > second:
            second = v
        else:
            pass
    sv = f"{first}{second}"
    iv = int(sv)
    answer += iv
print(answer)
```

```python
def get_jotage(bank:str, num_batteries:int) -> int:
    batteries = [-1]
    for b in range(1, num_batteries+1):
        batteries.append(batteries[b-1]+1)
        for i in range(batteries[b], len(bank) - num_batteries + b):
            if bank[i] > bank[batteries[b]]:
                batteries[b] = i
    on_batteries = [bank[i] for i in batteries[1:]]
    return int(''.join(on_batteries))
```

```python
total = 0
for bank in INPUT_DATA:
    total += get_jotage(bank, 12)
print(total)
```
