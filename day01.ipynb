# RAW INPUT

```python
with open('ex.txt', 'r') as f:
    ex_lines = [line.rstrip() for line in f.readlines()]
    print(f"ex.txt has {len(ex_lines)} lines")
with open('input.txt', 'r') as f:
    in_lines = [line.rstrip() for line in f.readlines()]
    print(f"input.txt has {len(in_lines)} lines")
```

# PARSE INPUT

```python
# PARSE STRATEGY
def parse(data):
    return [(line[0], int(line[1:])) for line in data]
```

```python
ex_data = parse(ex_lines)
in_data = parse(in_lines)
```

# Part 1

```python
count = 0
pos = 50
for dir, amt in in_data:
    pos += amt if dir == 'R' else -1 * amt
    pos %= 100

    if pos == 0:
        count += 1

print(count)
```

# Part 2

```python
count = 0
pos = 50
for dir, val in ex_data:
    for _ in range(val):
        if dir == 'R':
            pos += 1
        else:
            pos -= 1

        if pos == -1 or pos == 100:
            pos %= 100
        
        if pos == 0:
            count += 1

print(count)
```
