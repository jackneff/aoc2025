# RAW INPUT

```python
with open('ex.txt', 'r') as f:
    ex_lines = [line.rstrip() for line in f.readlines()]
    print(f"ex.txt has {len(ex_lines)} lines")
with open('input.txt', 'r') as f:
    in_lines = [line.rstrip() for line in f.readlines()]
    print(f"input.txt has {len(in_lines)} lines")
```

# PARSE INPUT

```python
# PARSE STRATEGY
def parse(data):
    data = data[0].split(",")
    data = [item.split('-') for item in data]
    return [(int(a), int(b)) for a,b in data]
```

```python
ex_data = parse(ex_lines)
in_data = parse(in_lines)
```

# Part 1

```python
answer = 0
for min, max in in_data:
    for v in range(min, max+1):
        sv = str(v)
        length = len(sv)
        if length % 2 != 0:
            continue 
        half = length // 2
        if sv[:half] == sv[half:]:
            #print(v)
            answer += v

print(answer)
```
