# IMPORT

```python
def import_lines(file):
    with open(file, 'r') as f:
        lines = [line.rstrip() for line in f.readlines()]
        return lines

EXAMPLE_LINES = import_lines('ex.txt')
print(f"ex.txt has {len(EXAMPLE_LINES)} lines")
INPUT_LINES = import_lines('input.txt')
print(f"input.txt has {len(EXAMPLE_LINES)} lines")
```

# PARSE

```python
# PARSE STRATEGY
def parse(data):
    data = data[0].split(",")
    data = [item.split('-') for item in data]
    return [(int(a), int(b)) for a,b in data]

EXAMPLE_DATA = parse(EXAMPLE_LINES)
INPUT_DATA = parse(INPUT_LINES)
LIMIT = 5
print(f"Example Data Top {LIMIT}: ")
print(EXAMPLE_DATA[:LIMIT])
print(f"Input Data Top {LIMIT}: ")
print(INPUT_DATA[:LIMIT])
```

# PART 1

```python
answer = 0
for min, max in INPUT_DATA:
    for v in range(min, max+1):
        sv = str(v)
        length = len(sv)
        if length % 2 != 0:
            continue 
        half = length // 2
        if sv[:half] == sv[half:]:
            #print(v)
            answer += v
print(answer)
```

# PART 2

```python
def split_words_into_chunks(words, chunk_size):
    """ Divides a string into chunks.  Sentance becomes words.  Word becomes strings. """
    return [words[i:i + chunk_size] for i in range(0, len(words), chunk_size)]
def all_equal(chunks):
    """ Returns true if all elements in list are equal (the same) """
    return len(set(chunks)) == 1
```

```python
MIN_SET_LEN = 2
answer = 0
for min, max in INPUT_DATA:
    for v in range(min, max+1):
        sv = str(v)
        sl = len(sv)
        if sl >= MIN_SET_LEN:
            for n in range(1, sl):
                if sl % n == 0:
                    chunks = [sv[i:i+n] for i in range(0, len(sv), n)]
                    if all_equal(chunks):
                        #print(f"{sv} is invalid")
                        answer += v
                        break
print(f"Answer: {answer}")
```
